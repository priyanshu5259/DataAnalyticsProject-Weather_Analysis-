```python
#We need to predict the temperature of tommorrow by using the temperature of today
data["target"]=data.shift(-1)["Temp_C"]  #By using this we have made temp to shift up
```

```python
data=data.iloc[:-1,:].copy()
```

```python
from sklearn.linear_model import Ridge
reg=Ridge(alpha=.1)
```

```python
predictors=["Temp_C","Dew Point Temp_C","Rel Hum_%","Wind Speed_km/h","Visibility_km","Press_kPa"]
```

```python
train=data.loc[:"2012-11-30"]
```

```python
test=data.loc["2012-12-01":]
```

```python
reg.fit(train[predictors],train["target"])
```

```python
predictions=reg.predict(test[predictors])
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(test["target"],predictions)
```

```python
combined=pd.concat([test["target"],pd.Series(predictions,index=test.index)],axis=1)
```

```python

def create_predictions(predictions,data,reg):
  train=data.loc[:"2012-11-30"]
  test=data.loc["2012-12-01":]
  reg.fit(train[predictors],train["target"])
  predictions=reg.predict(test[predictors])
  error=mean_absolute_error(test["target"],predictions)
  combined=pd.concat([test["target"],pd.Series(predictions,index=test.index)],axis=1)
  combined.columns=["actual","predictions"]
  return error,combined
```
